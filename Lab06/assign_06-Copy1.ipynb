```python
import csv
import numpy as np

# high_school_2019.csv 파일을 읽어들인다.
with open('high_school_2019.csv', 'r', encoding='utf-8') as f:
    reader = csv.reader(f)

    # 2차원 리스트를 생성한다.
    data = [[]]

    for row in reader:
        data.append(row)

    # 첫 번째 행은 빈 리스트이므로 삭제
    data = data[1:]

    # 숫자 데이터에 해당되는 열 인덱스를 추출한다.
    target_cols = [2, 3, 4, 5, 6, 7]

    # 각 행의 길이를 동일하게 만들기 위해 빈 문자열로 채워준다.
    for row in data:
        while len(row) < max(target_cols) + 1:
            row.append('')

    # 숫자 데이터를 ndarray에 저장한다. 첫 번째 행을 제외하고 생성
    data_array = np.array([[int(value.split()[0]) if value.split()[0].isdigit() else 0 for i, value in enumerate(row) if i in target_cols] for row in data[1:]])

    # ndarray를 출력한다.
    print(data_array)
```

```python
#Q1-3.전국의 고등학교에 대해 각 학교의 총 학생수(1,2,3학년 학생수의 합)를 구해 마지막 열에 삽입하여 출력하는 프로그램을 작성하시오.
# pandas는 사용하지 않고 numpy 함수 및 배열만 이용하여 코드를 작성하시오.

import csv
import numpy as np

# high_school_2019.csv 파일을 읽어들인다.
with open('high_school_2019.csv', 'r', encoding='utf-8') as f:
    reader = csv.reader(f)

    # 2차원 리스트를 생성한다.
    data = [[]]

    for row in reader:
        data.append(row)

    # 첫 번째 행은 빈 리스트이므로 삭제
    data = data[1:]

    # 숫자 데이터에 해당되는 열 인덱스를 추출한다.
    target_cols = [2, 3, 4, 5, 6, 7]

    # 각 행의 길이를 동일하게 만들기 위해 빈 문자열로 채워준다.
    for row in data:
        while len(row) < max(target_cols) + 1:
            row.append('')

    # 각 학교의 총 학생수를 계산하여 마지막 열에 삽입한다.
    for row in data:
        total_students = sum(int(value.split()[0]) if value.split()[0].isdigit() else 0 for i, value in enumerate(row) if i in target_cols)
        row.append(total_students)

    # 숫자 데이터를 ndarray에 저장한다.
    data_array = np.array(data[1:], dtype=object)  # dtype=object를 사용하여 모든 데이터 유형을 유지

    # ndarray를 출력한다.
    print(data_array)
```

```python
#Q1-4. 전국 고등학교에 대해 각 학년별 총 학생수를 출력하고, 막대 그래프를 통해 시각화하는 프로그램을 작성하시오.
# pandas는 사용하지 않고 numpy 함수 및 배열만 이용하여 코드를 작성하시오.

import csv
import numpy as np
import matplotlib.pyplot as plt

# high_school_2019.csv 파일을 읽어들인다.
with open('high_school_2019.csv', 'r', encoding='utf-8') as f:
    reader = csv.reader(f)

    # 2차원 리스트를 생성한다.
    data = [[]]

    for row in reader:
        data.append(row)

    # 첫 번째 행은 빈 리스트이므로 삭제
    data = data[1:]

    # 숫자 데이터에 해당되는 열 인덱스를 추출한다.
    target_cols = [2, 3, 4, 5, 6, 7]

    # 각 행의 길이를 동일하게 만들기 위해 빈 문자열로 채워준다.
    for row in data:
        while len(row) < max(target_cols) + 1:
            row.append('')

    # 각 학교의 총 학생수를 계산하여 마지막 열에 삽입한다.
    for row in data:
        total_students = sum(int(value.split()[0]) if value.split()[0].isdigit() else 0 for i, value in enumerate(row) if i in target_cols)
        row.append(total_students)

    # 숫자 데이터를 ndarray에 저장한다.
    data_array = np.array(data[1:], dtype=object)  # dtype=object를 사용하여 모든 데이터 유형을 유지

    # 각 학년별 총 학생수를 계산한다.
    total_students_by_grade = np.sum(data_array[:, [2, 4, 6]].astype(int), axis=0)

    # 막대 그래프를 통해 시각화한다.
    grades = ['1st Grade', '2nd Grade', '3rd Grade']
    plt.bar(grades, total_students_by_grade)
    plt.title('The total number of high school students nationwide.')
    plt.show()

# Q1. End
```
